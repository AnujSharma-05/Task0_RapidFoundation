```python
import time

def timer(func):
    def wrapper(*args, **kwargs): #here we are using *args and **kwargs to allow the wrapper function to accept any number of positional and keyword arguments, which will be passed to the original function when it is called.
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__}: Execution time: {end - start}") 

        #__name__ is a special attribute of a function that contains the name of the function as a string. By using func.__name__, we can access the name of the original function that is being decorated and include it in the print statement for better readability and debugging purposes.
        return result
    return wrapper



@timer #this is a decorator which make the below function to pass through the pipeline of the timer function before execution
def example_func(n):
    time.sleep(n)

example_func(2)
```

```python
#----Debugging Function Calls--------

def debugger(func):
    def wrapper(*args, **kwargs):
        print(f"\nArguments: {args}, {kwargs}")
        print(f"{func.__name__} is the function called")
        result = func(*args, **kwargs)
        print(f"Result: {result}")
        return result
    return wrapper





@debugger
def greet(name, greeting = "hellllo"):
    return(f"{greeting}, {name}!")

print(greet("Anuj"))
# print(greet("Anuj", greeting = "hii"))


@debugger
def another_func(a, b):
    return a + b

print(another_func(3, 4))

```

```python
#----------Caching Function Results----------
#Decorator to return the cached value if the function is called with the same arguments again, instead of recomputing the result.

def cache(func):
    cache_value = {}
    def wrapper(*args,**kwargs):
        print(cache_value)
        if args in cache_value:
            print(f"Using cached value for {func.__name__}{args}")
            return cache_value[args]
        result = func(*args,**kwargs)
        cache_value[args] = result
        return result
    return wrapper

@cache
def long_running_func(a,b):
    time.sleep(4)  # Simulate a long-running operation
    return a + b

print(long_running_func(3, 4))  
print(long_running_func(3, 4))  
print(long_running_func(3, 1))  
print(long_running_func(3, 1))  



```
