## Class definitions
The `Car` class demonstrates `__init__`, `self`, and an instance method. `ElectricCar` shows inheritance and `super()` usage.

```python
class Car:
    def __init__(self, brand, model):
        # This constructor initializes instance attributes.
        self.brand = brand
        self.model = model
    def full_name(self):
        return f"{self.brand} {self.model}"

my_car = Car("toyota","corolla")
print(my_car)  
print(my_car.model)  # Output: corolla
print(my_car.brand)  # Output: toyota

my_new_car = Car("honda","civic")
print(my_new_car.model)  # Output: civic
print(my_new_car.brand)  # Output: honda
print(my_new_car.full_name())  # Output: honda civic
```

```python
#--------------Inheritance---------------

class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model) #super() is used to call the __init__ method of the parent class (Car) to initialize the brand and model attributes. This allows the ElectricCar class to inherit the properties of the Car class while also adding its own unique attribute, battery_size.
        self.battery_size = battery_size


my_tesla = ElectricCar("tesla", "model S", "75 kWh")
print(my_tesla.full_name())  # Output: tesla model S
print(my_tesla.battery_size)  # Output: 75 kWh

```
