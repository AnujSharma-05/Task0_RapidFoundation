## Class definitions
The `Car` class demonstrates `__init__`, `self`, and an instance method. `ElectricCar` shows inheritance and `super()` usage.

```python
class Car:
    def __init__(self, brand, model):
        # This constructor initializes instance attributes.
        self.brand = brand
        self.model = model
    def full_name(self):
        return f"{self.brand} {self.model}"

my_car = Car("toyota","corolla")
print(my_car)  
print(my_car.model)  # Output: corolla
print(my_car.brand)  # Output: toyota

my_new_car = Car("honda","civic")
print(my_new_car.model)  # Output: civic
print(my_new_car.brand)  # Output: honda
print(my_new_car.full_name())  # Output: honda civic
```

```python
#--------------Inheritance---------------

class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model) #super() is used to call the __init__ method of the parent class (Car) to initialize the brand and model attributes. This allows the ElectricCar class to inherit the properties of the Car class while also adding its own unique attribute, battery_size.
        self.battery_size = battery_size


my_tesla = ElectricCar("tesla", "model S", "75 kWh")
print(my_tesla.full_name())  # Output: tesla model S
print(my_tesla.battery_size)  # Output: 75 kWh

```

```python
#--------------Encapsulation----------------
#it refers to according to first principles that we should hide the internal state and functionality of an object and only expose a public interface. This is typically achieved by using private attributes and methods, which are not accessible from outside the class. In Python, we can indicate that an attribute or method is intended to be private by prefixing its name with an underscore (_). This is a convention that signals to other developers that these members should not be accessed directly.


#use of private variables -> add __ before the variable name to make it private. This will trigger name mangling, which means that the variable will be renamed internally to include the class name, making it less accessible from outside the class.
class Car2:
    def __init__(self, brand, model, year):
        self.__brand = brand
        self.__model = model
        self.__year = year
    def get_brand(self):
        return self.__brand+"hehehe, this is an excapsulated and abstracted text"

class testcar(Car2):
    def __init__(self, brand, model, year):
        super().__init__(brand, model, year)
        

my_test_car = testcar("ford", "mustang", 2020)
print(my_test_car.get_brand())  # Output: fordhehehe, this is an excapsulated and abstracted text
print(my_test_car.__brand)  # This will raise an AttributeError because __brand is a private attribute and cannot be accessed directly from outside the class.

```

```python
#--------------Polymorphism----------------
#Polymorphism is the ability of methods to work with multiple behaviors. 
#can be achieved through method overriding, where a subclass provides a specific implementation of a method that is already defined in its superclass. This allows the same method name to be used for different types of objects, and the appropriate method will be called based on the type of object that is invoking it.


class Car:
    total_cars = 0
    def __init__(self, brand, model):
        # This constructor initializes instance attributes.
        self.brand = brand
        self.model = model
        # type(self).total_cars += 1
        Car.total_cars += 1

    def full_name(self):
        return f"{self.brand} {self.model}"

    def fuel_type(self):
        return "Petrol/diesel"


class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model)
        self.battery_size = battery_size

    def fuel_type(self):
        return "Electric"

my_car = Car("toyota","corolla")
print(my_car.fuel_type())  
my_tesla = ElectricCar("tesla", "model S", "75 kWh")
print(my_tesla.fuel_type())  



print(Car.total_cars)  # Output: 2

#This is runtime polymorphism because the method that gets called is determined at runtime based on the type of object (Car or ElectricCar) that is invoking the fuel_type() method.

```

```python
#--------------Static Methods----------------
# A method which is available to the class but not to its instances. 
# It does not take self or cls as the first parameter and cannot access instance or class variables. It is used for utility functions that are related to the class but do not require access to instance or class data.


class Car:
    total_cars = 0
    def __init__(self, brand, model):
        # This constructor initializes instance attributes.
        self.brand = brand
        self.model = model
        # type(self).total_cars += 1
        Car.total_cars += 1

    def full_name(self):
        return f"{self.brand} {self.model}"

    def fuel_type(self):
        return "Petrol/diesel"

    @staticmethod
    def general_info():
        return "Cars are AMAZINGGGGG."


class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model)
        self.battery_size = battery_size

    def fuel_type(self):
        return "Electric"

my_car = Car("toyota","corolla")
print(my_car.fuel_type())  
my_tesla = ElectricCar("tesla", "model S", "75 kWh")
print(my_tesla.fuel_type()) 



print(Car.general_info())  # Output: Cars are AMAZINGGGGG. (if self is not passed to general_info method)

print(my_car.general_info())  # Output: ERROR -> 0 positional arguments required (if self is not passed to general_info method)

#Case -> when self is passed to general_info method
print(Car.general_info())  # Output: error, explain this Car.general_info() missing 1 required positional argument: 'self'. Answer -> This error occurs because the general_info method is defined to take a self parameter, which is typically used to refer to the instance of the class. When you call Car.general_info(), it is trying to call the method without providing an instance (self), which leads to the missing argument error. To fix this, you can either make general_info a static method (by adding @staticmethod decorator) or call it on an instance of the class instead of the class itself.

print(my_car.general_info())  # Output: Cars are AMAZINGGGGG. WORKS FINE!

# Now Static Decorator is added to general_info method
print(Car.general_info())  # Output: Cars are AMAZINGGGGG. (if self is not passed to general_info method)

print(my_car.general_info())  # this should not run as 


print(Car.total_cars)  # Output: 2
```

```python
#----------USE A PROPERTY DECORATOR AND MAKE THE MODEL ATTRIBUTE "READ ONLY" ---------


class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.__model = model  # Use a single underscore to indicate that this attribute is intended to be private
    
    @property
    def model(self):
        return self.__model #this is a getter method that uses the @property decorator to allow read-only access to the __model attribute. By defining a property, you can access the model attribute using my_car.model, but you cannot modify it directly (e.g., my_car.model = "new model" would raise an AttributeError). This way, the model attribute is effectively read-only from outside the class.


my_car = Car("toyota","corolla")
print(my_car.model)  # Output: corolla
# my_car.model = "classic"  #attribute error
```

```python
#--use of isinstance()
#to check if an object is an instance of a specific class or a subclass thereof.
print( isinstance(my_car, Car)  ) 
print (isinstance(my_tesla, ElectricCar))
print (isinstance(my_tesla, Car))
```

```python
#----------Multiple Inheritance--------------

class Battery:
    def battery_info(self):
        return "This is a battery."

class Engine:
    def engine_info(self):
        return "This is an engine."

class ElectricCarTwo(Battery, Engine, Car):
    pass

my_new_tesla = ElectricCarTwo("tesla","model s" )
print(my_new_tesla.battery_info())  
print(my_new_tesla.engine_info())
print(my_new_tesla.model)
```
